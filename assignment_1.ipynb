```python
#droping athletes column because cannot fill
df.drop(columns = "athletes", inplace = True)
```

```python
#droping country column because country_code is there
df.drop(columns = "country", inplace = True)
```

```python
df.columns
```

```python
#filling country_code with first mode
df['country_code'] = df['country_code'].fillna(df['country_code'].mode()[0])
df.head()
```

```python
df.isna().sum()
```

```python
#filling medal with mode
df['medal'] = df['medal'].fillna(df['medal'].mode()[0])
df.head()
```

```python
#removing duplicate values
df.drop_duplicates(inplace = True)
```

```python
df.sample(10)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#encoding categorical variables 
df["season"].replace({"Summer" : 0, "Winter" : 1}, inplace = True)
```

```python
df.head()
```

```python
df["medal"].replace({"Gold" : 0, "Silver" : 1, "Bronze" : 2}, inplace = True)
df["event_gender"].replace({"Men's" : 0, "Mixed" : 1, "Women's" : 2}, inplace = True)
```

```python
df.head()
```

```python
#normalizing using min-max 
df["year"] = (df["year"] - df["year"].min()) / (df["year"].max() - df["year"].min())
df.head()
```

```python
df["event_gender"] = (df["event_gender"] - df["event_gender"].min()) / (df["event_gender"].max() - df["event_gender"].min())
df.head()
```

```python
df["medal"] = (df["medal"] - df["medal"].min()) / (df["medal"].max() - df["medal"].min())
df.head()
```

```python
df["sport"].nunique()
```

```python
print("Total diff games : ",df["games"].nunique())
print("Total diff events conducted : ",df["event_name"].nunique())
print("Total diff countries participated : ",df["country_code"].nunique())
```

```python
#sd of numeric columns
df.std(numeric_only=True)
```

```python
#variance
df.var(numeric_only=True)
```

```python
#Range
df.max(numeric_only=True) - df.min(numeric_only=True)
```

```python
#outliers using IQR method
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```
